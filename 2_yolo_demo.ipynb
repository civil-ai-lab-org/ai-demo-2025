```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')

model.info()
```

```python
results = model.predict("detection1.jpg", imgsz=320)

results[0].show()

results[0].save(filename='result_imge1.jpg')
print('ファイルにセーブしました')
```
