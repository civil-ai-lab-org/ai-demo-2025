画像生成（Text-to-Image）

モデル例: stabilityai/stable-diffusion-2-1 (Stable Diffusion)

デモ内容:

入力したテキスト（例: 「ロボットのようなかわいい猫」）から画像を生成


```python
from google.colab import userdata
from diffusers import StableDiffusionPipeline
import torch
```

```python
# シークレットからトークンを読み込む
access_token = userdata.get('HUGGINGFACE_TOKEN')
```

```python
# モデルのロード
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, use_auth_token=access_token)
pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# 画像生成
#prompt = "a cute cat sleeping on a sofa"
#prompt = "a robot cat wearing a superhero cape, cute and futuristic"
#prompt = "a robot cat in a magical forest, glowing eyes, fantasy style"
#prompt = "a robot cat chef cooking in a tiny kitchen, cartoon style"
prompt = "a cute cat like a robot"

#"a cute cat sitting on a couch"
image = pipe(prompt).images[0]
image.save("generated_image.png")
image
```
